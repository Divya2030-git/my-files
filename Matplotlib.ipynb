#Line chart:

```python
import pandas as pd
```

```python
dates = pd.date_range(start = "2025-01-01", periods = 12, freq = "ME")
placement_data = [100, 90, 0, 20, 30, 50, 60, 80, 90, 35, 25, 95]
plt.plot(dates, placement_data, marker = 'o')
plt.xticks(dates, dates.strftime("%b"))
for x,y in zip(dates, placement_data):
    plt.text(x,y, str(y))
plt.grid(True)
plt.show()
```

# Multiple line chart:

```python
dates = pd.date_range(start = "2013-01-01", periods = 12, freq = "YE")
Prabhas_fans = [10, 20, 40, 55, 100, 35, 15, 0, 200, 300, 350, 400]
Mahesh_fans = [50, 60, 70, 20, 10, 0, 100, 150, 200, 400, 600, 1000]
Anushka_fans = [5, 10, 20, 30, 40, 50, 60, 70, 90, 110, 120, 150]

plt.plot(dates, Prabhas_fans, marker = 'o', label = "Prabhas")
plt.plot(dates, Mahesh_fans, marker = 's', label = "Mahesh")
plt.plot(dates, Anushka_fans, marker = 's', label = "Anushka")

plt.legend()
plt.show()
```

# Pie Chart:

```python
import matplotlib.pyplot as plt
```

```python
brands = ['Oppo', 'Samsung', 'apple', 'Redmi','IQOO','Other']
market_share = [10, 5, 30, 35, 8, 12]

plt.figure(figsize = (4, 4))
plt.pie(market_share, labels = brands, autopct = '%1.1f%%')
plt.title("Smartphone market share")
plt.savefig("Pie_chart123.pdf", dpi = 300, bbox_inches = 'tight')
plt.show()
                      
```

# Multiple subplot:

Method:1

```python
Dream_Thali = ['Mutton Curry', 'Pachi Pulusu', 'Paneer buttir masala', 'Biryani', 'Chicken drum stick', 'Kunufa']
share = [20, 10, 20, 30, 10, 10]


plt.figure(figsize = (10,5))
plt.subplot(2, 2, 1)
plt.pie(share, labels = Dream_Thali, autopct = '%1.1f%%')
plt.title('Sub-plot-1')

plt.subplot(2, 2, 2)
plt.pie(share, labels = Dream_Thali, autopct = '%1.1f%%')
plt.title('Sub-plot-2')


plt.subplot(2, 2, 3)
plt.pie(share, labels = Dream_Thali, autopct = '%1.1f%%')
plt.title('Sub-plot-3')

plt.subplot(2, 2, 4)
plt.pie(share, labels = Dream_Thali, autopct = '%1.1f%%')
plt.title('Sub-plot-4')
plt.tight_layout()
```

Method:2

```python
Dream_Thali = ['Mutton Curry', 'Pachi Pulusu', 'Paneer buttir masala', 'Biryani', 'Chicken drum stick', 'Kunufa']
share = [20, 10, 20, 30, 10, 10]

fig, axes = plt.subplots(2,2, figsize = (10, 10))
titles = ['Sub-plot-1', 'Sub-plot-2', 'Sub-plot-3', 'Sub-plot-4']
print(axes)

for ax, title in zip(axes.ravel(), titles):
    ax.pie(share, labels = Dream_Thali, autopct = "%1.1f%%")
    ax.set_title(title)    
```

# 3D Charts:

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
z = np.linspace(0, 20, 100)
x = np.sin(z)
y = np.cos(z)

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

ax.plot(x,y,z)
```
