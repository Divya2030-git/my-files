9.Time Series related Functions

```python
date_rng = pd.date_range(start = '2020-01-01', end = '2025-12-31', freq = 'h')
date_rng
```

```python
# selecting 1000 records from above date time data
import numpy as np
```

```python
np.random.seed(42)
random_dates = np.random.choice(date_rng, size = 1000, replace = False)
print(len(random_dates))
random_dates
```

```python
df = pd.DataFrame({
    'timestamp' : pd.to_datetime(random_dates)
})
df
    
```

```python
df = pd.DataFrame({
    'timestanp' : pd.to_datetime(random_dates),
    'order_id' : np.arange(1, 1001),
    'category' : np.random.choice(['Biryani', 'Paya', 'Irani', 'Chai', 'Maska Bun'], size = 1000),
    'Quantity' : np.random.randint(1,5, size = 1000),
    'Price' : np.random.randint(25, 450, size = 1000)
})
df
    
```

```python
df['Sales'] = df['Price'] * df['Quantity']
df
```

```python
df = df.sort_values('timestanp').reset_index(drop = True)
df
```

```python
df['Year'] = df.timestanp.dt.year
df
```

```python
df['Month'] = df.timestanp.dt.month
df
```

```python
df['Day'] = df.timestanp.dt.day
df
```

```python
df['Hour'] = df.timestanp.dt.hour
df
```

```python
df['Minute'] = df.timestanp.dt.minute
df
```

```python
df['Second'] = df.timestanp.dt.second
df
```
